```python
#import libraries

import datetime
import os
import win32com.client
import glob
from datetime import date
import timeit

#folder paths

rS = timeit.default_timer()

path_completed_approvals = r"C:\Users\tiftor\Vships\BPM_Global_CPI_Reports - BPM Repo - BPM Repo\HR ATR Approval Report\Completed Approval Timeline Trends"
path_approved_curr_stat = r"C:\Users\tiftor\Vships\BPM_Global_CPI_Reports - BPM Repo - BPM Repo\HR ATR Approval Report\Approved Vacancies & Current Status"
path_time_to_hire = r"C:\Users\tiftor\Vships\BPM_Global_CPI_Reports - BPM Repo - BPM Repo\HR ATR Approval Report\Time to Hire"
path_pending_approval = r"C:\Users\tiftor\Vships\BPM_Global_CPI_Reports - BPM Repo\HR ATR Approval Report\Pending Approvals"
today = datetime.date.today()

#remove files on folders

fn1 = glob.glob(path_completed_approvals + "/*.csv")
for a in fn1:
    os.remove(a)
    
fn2 = glob.glob(path_approved_curr_stat + "/*.csv")
for b in fn2:
    os.remove(b)
    
fn3 = glob.glob(path_time_to_hire + "/*.csv")
for c in fn3:
    os.remove(c)
    

fn4 = glob.glob(path_pending_approval + "/*.csv")
for d in fn4:
    os.remove(d)

fn5 = glob.glob(path_time_to_hire + "/*.xlsx")
for e in fn5:
    os.remove(e)    
    
#get attachments from outlook

outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
inbox = outlook.GetDefaultFolder(6)
messages = inbox.Items
#last_message = messages.Getlast()


#save attachments to local folder (completed approvals)

def save_attachments(subject_prefix):
    messages.Sort("[ReceivedTime]", True)
    for message in messages:
        if message.Subject.startswith(subject_prefix):
            print("saving attachments for:", message.Subject)
            for attachment in message.Attachments:
                print(attachment.FileName)
                attachment.SaveAsFile(os.path.join(path_completed_approvals, str(attachment.FileName)))
            return
                

if __name__ == "__main__":
    save_attachments('Report results (Completed Approval Timeline Trends)')
    
    
    
#save attachments to local folder (approved&current status)

def save_attachments(subject_prefix):
    messages.Sort("[ReceivedTime]", True)
    for message in messages:
        if message.Subject.startswith(subject_prefix):
            print("saving attachments for:", message.Subject)
            for attachment in message.Attachments:
                print(attachment.FileName)
                attachment.SaveAsFile(os.path.join(path_approved_curr_stat, str(attachment.FileName)))
            return
                
if __name__ == "__main__":
    save_attachments('Report results (Active Filled Archived on Hold Vacancies)')
    

#save attachments to local folder (time to hire)
def save_attachments(subject_prefix):
    messages.Sort("[ReceivedTime]", True)
    for message in messages:
        if message.Subject.startswith(subject_prefix):
            print("saving attachments for:", message.Subject)
            for attachment in message.Attachments:
                print(attachment.FileName)
                attachment.SaveAsFile(os.path.join(path_time_to_hire, str(attachment.FileName)))
            return
                
if __name__ == "__main__":
    save_attachments('Report results (*Average Time to Hire)')
    
    
    
#save attachments to local folder (pending approvals)

def save_attachments(subject):
    for message in messages:
        if message.Subject == subject:
            for attachment in message.Attachments:
                print(attachment.FileName)
                attachment.SaveAsFile(os.path.join(path_pending_approval, str(attachment)))
                
if __name__ == "__main__":
    save_attachments('Report results (ATR <6 MTHS -Pending Business Approver)')
    

rE = timeit.default_timer()
print ('Time: ', rE - rS)
```
