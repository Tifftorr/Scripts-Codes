```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, Alignment
import timeit

rS = timeit.default_timer()

file_path = 'C:\\Users\\tiftor\\Vships\\Operational Excellence - BPM (1)\\Executive Reports\\BU Risks\\BU Risk Register.xlsx'
sheet_name = 'Singapore'  # Name of the new sheet

try:
    # Load the existing workbook
    book = load_workbook(file_path)
    
    # Check if the sheet_name already exists, and if so, remove it
    if sheet_name in book.sheetnames:
        del book[sheet_name]
    
    # Save the workbook to apply the changes
    book.save(file_path)

    # Create a Pandas Excel writer using openpyxl as the engine
    with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
        # Assign the existing workbook to the writer
        writer.book = book
        
        # Copy existing sheets to the writer
        writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
        
        # Assuming 'concat_months' is your DataFrame that you want to write to the new sheet
        concat_months.to_excel(writer, sheet_name=sheet_name, index=False)

    # After saving through pandas, reload the workbook to modify properties
    book = load_workbook(file_path)
    sheet = book[sheet_name]

    # Set column width
    for col in sheet.columns:
        sheet.column_dimensions[col[0].column_letter].width = 20

    # Apply font, alignment, and text wrap to all cells
    for row in sheet.iter_rows():
        for cell in row:
            cell.font = Font(name="Amina", size=8)  # Adjust the font name as necessary
            # Here we add wrapText=True to ensure text wrapping
            cell.alignment = Alignment(horizontal="center", vertical="center", wrapText=True)

    # Save the changes to the workbook
    book.save(file_path)

except Exception as e:
    print(f"An error occurred: {e}")
    
rE = timeit.default_timer()
print ('Time: ', rE - rS)
```
